```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pytz
```

```python
url = 'https://www.flickr.com/search/people/?username=david+anderson'
```

```python
response = requests.get(url)
```

```python
import json

data = json.loads(response.text.split('modelExport:')[1].split('auth')[0].strip(',\n\t\t\t\t'))

with open('data.json', 'w') as file:
	json.dump(data, file, indent=4)
```

```python
combined_data = list()
main = data['main']['search-contacts-models'][0]['data']['contacts']['data']['_data']
for user_data in main:
	user = dict()
	
	profile_user = user_data['data']
	user['username'] = profile_user['username']
	user['realname'] = profile_user['realname']
	user['image'] = profile_user['buddyicon']['data']['retina']
	try:
		user['follower'] = profile_user['followersCount']
	except:
		user['follower'] = ''
	
	try:
		user['image_count'] = profile_user['photosCount']
	except:
		user['image_count'] = ''
	
	joined_date = int(profile_user['dateJoined']['data'])
	user['joined_date'] = datetime.fromtimestamp(joined_date).strftime('%b %Y')

	combined_data.append(user)
combined_data
		
```

```python
import csv

with open('flickr_data.json', 'w') as file:
    json.dump(combined_data, file, indent=4)
```
